Adding a new weapon category column

```python
firearm_search = ['Handgun - pistol, revolver, etc','Shotgun','Firearm, type not stated', 'Rifle','Other gun']
knife_search = ['Knife or cutting instrument']
bluntobject_search = ['Blunt object - hammer, club, etc']
personal_search = ['Personal weapons, includes beating']
other_search = ['Drowning','Fire','Asphyxiation - includes death by gas','Narcotics or drugs, sleeping pills',
                'Strangulation - hanging','Explosives', 'Poison - does not include gas','Other or type unknown','Pushed or thrown out window']

def weapon_used(df):
    
    if df['Weapon'] in firearm_search:
        return 'Firearm'
    elif df['Weapon'] in knife_search:
        return 'Knife'
    elif df['Weapon'] in bluntobject_search:
        return 'Blunt object'
    elif df['Weapon'] in personal_search:
        return 'Personal'
    elif df['Weapon'] in other_search:
        return 'Other'

df['WeaponCatg'] = df.apply(weapon_used, axis = 1)
```

Adding a new relationship category column

```python
family_search = ['Husband','Wife','Common-law husband','Common-law wife','Mother', 'Father','Son',
                'Daughter','Brother','Sister','In-law','Stepfather','Stepmother','Stepson','Stepdaughter','Other family']
acquaintance_search = ['Acquaintance','Boyfriend','Girlfriend','Ex-husband','Ex-wife','Employee','Employer',
                       'Friend','Homosexual relationship','Neighbor','Other - known to victim']
stranger_search = ['Stranger']
unknown_search = ['Relationship not determined']

 

def relationship(df):
    
    if df['Relationship'] in family_search:
        return 'Family'
    elif df['Relationship'] in acquaintance_search:
        return 'Acquaintance'
    elif df['Relationship'] in stranger_search:
        return 'Stranger'
    elif df['Relationship'] in unknown_search:
        return 'Unknown'

df['RelationshipCatg'] = df.apply(relationship, axis = 1)
```

```python
homicide_status = df.groupby('Solved').sum().reset_index()[['Solved','Count']]
homicide_status['percentage'] = (homicide_status['Count']/homicide_status['Count'].sum()) * 100
homicide_status['percentage'] = homicide_status['percentage'].round(decimals = 1)
homicide_status
```
